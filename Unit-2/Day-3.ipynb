```python
"""---Choropleth Maps---"""

import folium
import pandas as pd

state_unemp = pd.read_csv('US_Unemployment_Oct2012.csv')
state_geo = 'us-states.json'

usa_state = folium.Map(location=[48,-102], zoom_start=4)

folium.Choropleth(geo_data=state_geo, name='choropleth', data=state_unemp, columns=['State', 'Unemployment'], key_on='feature.id', fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2, legend_name='Unemployment Scale').add_to(usa_state)

usa_state
```

```python
"""---NetworkX---"""

# 1.) Undirected Graph

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5,6]) # OR "G.add_node(1)" OR "G.add_nodes_from(range(1,7))"

G.add_edge(1,2)
G.add_edge(1,1) # OR "G.add_edges_from([(1,2),(1,1),(2,3),(3,6),(4,6),(5,6)])"
nx.draw_networkx(G, node_size=850, node_color='red', width=2, edge_color='black')
plt.show()
```

```python
# 2.) Directed Graph

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from([1,2,3,4,5,6])
G.add_edges_from([(1,2),(1,1),(2,3),(3,6),(4,6),(5,6)])
nx.draw_networkx(G, node_size=850, node_color='red', width=2, edge_color='black')
plt.show()
```

```python
# Create a Directed Graph from adjacency matrix

import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_nodes_from(range(4))
l = [[0,1,0,1],[0,0,0,1],[1,0,0,0],[1,0,1,0]]

for i in range(4):
    for j in range(4):
        if l[i][j]==1:
            G.add_edge(i,j)
        
nx.draw_networkx(G, node_size=850, node_color='red', width=2, edge_color='black')
plt.show()
```
