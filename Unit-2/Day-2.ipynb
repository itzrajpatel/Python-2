```python
"""---Word Cloud---"""
#Words that are repeating maximum times will be in large fonts and words that are repeating mainium times will be in small fonts
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

alice_novel = open('alice_in_wonderland.txt')
alice_novel = alice_novel.read()
stopwords = set(STOPWORDS)
#print(len(stopwords)) --> 192

alice_wc = WordCloud(background_color='black', max_words=2000,stopwords=stopwords)
stopwords.add('said') # this will remove the particuar word from word cloud and add word in STOPWORDS
stopwords.add('Alice')
print(len(stopwords)) # --> 194
alice_wc.generate(alice_novel)
plt.imshow(alice_wc)
plt.axis('off')
plt.show()
```

```python
import folium
import pandas as pd

data = pd.read_csv('Police_Department_Incidents_-_Previous_Year__2016_.csv')
data.shape
```

```python
data.head()
```

```python
df=data.iloc[0:100] # Or "df=data.loc[0:99]"
```

```python
df.shape
```

```python
worldmap = folium.Map(location=[37.7749,-122.4194],zoom_start=10)
worldmap
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng in zip(df.Y,df.X):
    incidents.add_child(folium.CircleMarker([lat,lng], radius=5, color='black', fill=True, fill_color='cyan', fill_opacity=0.6))

worldmap.add_child(incidents)
```

```python
incidents = folium.map.FeatureGroup()
for lat,lng,labels in zip(df.Y, df.X, df.Category):
    incidents.add_child(folium.CircleMarker([lat,lng], radius=5, color='black', fill=True, fill_color='red', fill_opacity=0.6)),folium.Marker([lat,lng], popup=labels).add_to(worldmap)

worldmap.add_child(incidents)
```
