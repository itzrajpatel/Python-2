```python
import pandas as pd

df = pd.read_csv("FuelConsumptionCo2.csv")
df
```

```python
df['Age'] = 2024-df['MODELYEAR']
df
```

```python
df.drop(columns=['MODELYEAR','MAKE', 'MODEL', 'VEHICLECLASS', 'TRANSMISSION'], axis=1,inplace=True)
df
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
df.columns
```

```python
df.corr()
```

```python
x = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG','FUELCONSUMPTION_CITY','FUELTYPE_E','FUELTYPE_X','FUELTYPE_Z']]
y = df['CO2EMISSIONS']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x_train, y_train)
y_pred = lm.predict(x_test)
print(y_pred)
```

```python
diff = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
diff
```

```python
print(x.columns)
print('Coefficients:',lm.coef_)
print('Intercept:',lm.intercept_)
```

```python
from sklearn import metrics

print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score:", metrics.r2_score(y_test, y_pred))
```

```python
diff = pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
diff
```
