```python
"""---Filling Dummy Data---"""

import pandas as pd
import numpy as np

data = {'name':['William','Enna','Sofia','marcus','Edward','Thomas','Ethan',np.nan,'Arun','Anika'],
        'region':[np.nan,'North','East',np.nan,np.nan,'West','South','South',np.nan,'East'],
        'sales':[50000,52000,np.nan,np.nan,42000,72000,49000,np.nan,67000,65000],
        'expenses':[42000,43000,np.nan,np.nan,38000,39000,42000,np.nan,39000,50000]}
sales_data = pd.DataFrame(data)
sales_data
```
